```python
import random

# Word list for the game
word_list = ["python", "hangman", "programming", "game", "mindgame", "coding"]

# Function to display the word with guessed letters
def display_word(word, guessed_letters):
    display_word = ""
    for letter in word:
        if letter in guessed_letters:
            display_word += letter  # Show guessed letter
        else:
            display_word += "_"  # Hide unguessed letter
    return display_word  # Return completed word display

# Choose a random word
word = random.choice(word_list).lower()
guessed_letters = set()
attempts = 8  # Maximum attempts

# Game loop
while attempts > 0:
    print("\nğŸ® WELCOME TO HANGMAN GAME ğŸ­")  # Added game icon in heading
    print(" Guess the hidden word before you run out of attempts! ğŸ’¡")
    print(f"Word: {display_word(word, guessed_letters)}")  # Display word progress
    print(f"ğŸ“ Guessed Letters: {', '.join(guessed_letters) if guessed_letters else 'None'}")
    print(f"â¤ï¸ Attempts Left: {attempts}")

    # Get user input
    guess = input("ğŸ”¤ Enter a Letter: ").lower()

    # Check if input is valid
    if len(guess) != 1 or not guess.isalpha():
        print("âš ï¸ Invalid input! Please enter a single letter.")
        continue  # Skip to next iteration

    # Check if letter was already guessed
    if guess in guessed_letters:
        print("ğŸ” You already guessed that letter! Try a different one.")
        continue

    guessed_letters.add(guess)  # Add guess to set

    # Check if guess is correct
    if guess in word:
        print("âœ… Good job! That letter is in the word.")
    else:
        print("âŒ Wrong guess! Try again.")
        attempts -= 1  # Reduce attempts for wrong guess

    # Check if player has guessed the entire word
    if "_" not in display_word(word, guessed_letters):
        print(f"\nğŸ‰ Congratulations! You guessed the word: {word} ğŸ‰")
        break

# If attempts run out
if attempts == 0:
    print(f"\nğŸ’€ Game Over! The correct word was: {word} ğŸ’€")
```
