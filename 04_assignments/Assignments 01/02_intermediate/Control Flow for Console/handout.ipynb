```python
import random

# Milestone #5: Adding a points system
NUM_ROUNDS = 5

# Milestone #1: Generate the random numbers
# Function to generate random numbers for the player and the computer
def generate_numbers():
    player_number = random.randint(1, 100)  # Player's random number between 1 and 100
    computer_number = random.randint(1, 100)  # Computer's random number between 1 and 100
    return player_number, computer_number  # Return both numbers

# Milestone #2: Get the user choice and game logic
# Function to play a single round of the game
def play_round(player_number, computer_number):
    print(f"Your number is {player_number}")  # Show the player their number

    # Ask the user if they think their number is higher or lower than the computer's number
    guess = input("Do you think your number is higher or lower than the computer's?: ").lower()

    # Milestone #4: Safeguard user input - Ensure the input is either 'higher' or 'lower'
    while guess not in ["higher", "lower"]:
        guess = input("Please enter either 'higher' or 'lower': ").lower()

    # Milestone #3: Game logic
    # Check if the player's guess matches the actual comparison
    if (guess == "higher" and player_number > computer_number) or (guess == "lower" and player_number < computer_number):
        print(f"You were right! The computer's number was {computer_number}")
        return True  # Correct guess
    elif player_number == computer_number:
        print(f"The computer's number was {computer_number}")
        print("Aww, that's incorrect.")  # Same numbers, not higher or lower
        return False  # Incorrect guess (same numbers, so it's a loss)
    else:
        print(f"Aww, that's incorrect. The computer's number was {computer_number}")
        return False  # Incorrect guess

# Milestone #4: Play multiple rounds of the game
# Function to play the full game with multiple rounds
def play_game():
    score = 0  # Initialize the score to 0
    print("Welcome to the High-Low Game!")  # Welcome message
    print("--------------------------------")

    # Loop through all the rounds
    for round_number in range(1, NUM_ROUNDS + 1):  # Iterating through rounds (1 to NUM_ROUNDS)
        print(f"Round {round_number}")  # Print round number

        player_number, computer_number = generate_numbers()  # Generate random numbers for the round
        is_correct = play_round(player_number, computer_number)  # Play the round and get the result

        if is_correct:  # If the player's guess is correct, increase their score
            score += 1

        print(f"Your score is now {score}\n")  # Display the player's score after each round

    # Milestone #5: Adding a points system - Game over and performance evaluation
    print("Thanks for playing!")  # End of game message

    # Conditional messages based on player's performance
    if score == NUM_ROUNDS:
        print("Wow! You played perfectly!")  # Perfect game
    elif score >= NUM_ROUNDS // 2:
        print("Good job, you played really well!")  # At least half of the rounds won
    else:
        print("Better luck next time!")  # Less than half of the rounds won

# Start the game when the script runs
play_game()
```
